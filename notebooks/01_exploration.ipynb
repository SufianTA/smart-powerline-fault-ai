# Exploration: Synthetic Signals & Images
This notebook previews a few generated samples.


```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
from PIL import Image
df = pd.read_csv('data/synthetic/sample_signals.csv')
df.head()
```

```python
row = df.sample(1).iloc[0]
sig = np.load(row['signal_path'])
plt.figure(); plt.plot(sig); plt.title(f"Label={row['label']}")
img = Image.open(row['image_path']).convert('L')
display(img)
```
